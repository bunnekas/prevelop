```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# remove substring ' mm' from data in column 'Bohrungsdurchmesser [mm]'
cad_data['Bohrungsdurchmesser [mm]'] = cad_data['Bohrungsdurchmesser [mm]'].str.replace(' mm', '')
cad_data['Bohrungsdurchmesser [mm]'] = cad_data['Bohrungsdurchmesser [mm]'].str.replace(',', '.')
# convert data in column 'Bohrungsdurchmesser [mm]' to numeric
cad_data['Bohrungsdurchmesser [mm]'] = pd.to_numeric(cad_data['Bohrungsdurchmesser [mm]'])
```

```python
# remove substring ' mm' from data in column 'Bohrungstiefe [mm]'
cad_data['Bohrungstiefe [mm]'] = cad_data['Bohrungstiefe [mm]'].str.replace(' mm', '')
cad_data['Bohrungstiefe [mm]'] = cad_data['Bohrungstiefe [mm]'].str.replace(',', '.')
# for rows with substring '-' in column 'Bohrungstiefe [mm]', calculate the mean of the values before and after the '-'
# split the string at the '-' and convert the two values to numeric
cad_data['Bohrungstiefe [mm]'] = cad_data['Bohrungstiefe [mm]'].apply(lambda x: (float(x.split('-')[0]) + float(x.split('-')[1]))/2 if '-' in x else x)
# convert data in column 'Bohrungstiefe [mm]' to numeric
cad_data['Bohrungstiefe [mm]'] = pd.to_numeric(cad_data['Bohrungstiefe [mm]'])
```

```python
# load data
process_data= pd.read_excel(r"C:\Users\IM-KI\Documents\PrEvelOp_Daten_Ortlinghaus\PrEvelOp_231018_Fertigungsauftraege_Maschinendaten.xlsx")
```

```python
# concatenate columns 'Start Datum' and 'Start Uhrzeit' as strings to one column 'Start'
process_data['Start'] = process_data['Start Datum'].astype(str) + ' ' + process_data['Start Uhrzeit'].astype(str)
# convert data in column 'Start' to datetime
process_data['Start'] = pd.to_datetime(process_data['Start'], format='%Y-%m-%d %H:%M:%S')
```

```python
# concatenate columns 'Ende Datum' and 'Ende Uhrzeit' as strings to one column 'Ende'
process_data['Ende'] = process_data['Ende Datum'].astype(str) + ' ' + process_data['Ende Uhrzeit'].astype(str)
# convert data in column 'Ende' to datetime
process_data['Ende'] = pd.to_datetime(process_data['Ende'], format='%Y-%m-%d %H:%M:%S')
```

```python
# drop columns 'Start Datum', 'Start Uhrzeit', 'Ende Datum', 'Ende Uhrzeit'
process_data.drop(['Start Datum', 'Start Uhrzeit', 'Ende Datum', 'Ende Uhrzeit'], axis=1, inplace=True)
```

```python
# store unique values of column 'Fauf' in a list
unique_Fauf = process_data['Fauf'].unique()
```

```python
#drop rows in link_cad_process with values in column 'Fauf' that are not in unique_Fauf
link_cad_process = link_cad_process[link_cad_process['Fauf'].isin(unique_Fauf)]
```

```python
# drop rows in cad_data with values in column 'Matnr' that are not in link_cad_process[Artikelnummer]
cad_data = cad_data[cad_data['Matnr'].isin(link_cad_process['Artikelnummer'])]
```

```python
process_data_2 = pd.DataFrame(columns=['Fauf'])
process_data_2['Fauf'] = process_data['Fauf'].unique()
```

```python
for index, row in process_data_2.iterrows():
    # calculate the difference between the maximum and minimum value in column 'Start' for the selected rows
    process_data_2.loc[index, 'Duration'] = process_data[process_data['Fauf'] == row['Fauf']]['Ende'].max() - process_data[process_data['Fauf'] == row['Fauf']]['Start'].min()
    # calculate the average value in column 'PLAN-Bearbeitungszeit (Stück)' for the selected rows
    process_data_2.loc[index, 'Average processing time'] = process_data[process_data['Fauf'] == row['Fauf']]['PLAN-Bearbeitungszeit (Stück)'].mean()
    # calculate the average value in column 'PLAN-Rüstzeit (Los)' for the selected rows 
    process_data_2.loc[index, 'Average setup time'] = process_data[process_data['Fauf'] == row['Fauf']]['PLAN-Rüstzeit (Los)'].mean()
    # calculate the average value in column 'Losgröße' for the selected rows
    process_data_2.loc[index, 'Average batch size'] = process_data[process_data['Fauf'] == row['Fauf']]['Losgröße'].mean()
    # calulate the number of rows in process_data for the selected rows
    process_data_2.loc[index, 'Number of operations'] = len(process_data[process_data['Fauf'] == row['Fauf']])
```

```python
# convert data in column 'Duration' to seconds
process_data_2['Duration'] = process_data_2['Duration'].dt.total_seconds()
```

```python
# inner join of cad_data and link_cad_process on column 'Matnr' and 'Artikelnummer'
merged_data = pd.merge(cad_data, link_cad_process, left_on='Matnr', right_on='Artikelnummer', how='inner')
# inner join of merged_data and process_data_2 on column 'Fauf'
merged_data = pd.merge(merged_data, process_data_2, on='Fauf', how='inner')
```

```python
# make column 'Fauf' as key
merged_data.set_index('Fauf', inplace=True)
# drop columns 'Matnr', 'Artikelnummer'
merged_data.drop(['Matnr', 'Artikelnummer'], axis=1, inplace=True)
```
